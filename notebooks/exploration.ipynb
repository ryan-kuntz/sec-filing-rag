```python
from bs4 import BeautifulSoup
from pathlib import Path

with open("../data/raw/apple/10k_latest.html", "rb") as f:
    soup = BeautifulSoup(f, "html.parser")

# See what tags are actually being used
tags = [tag.name for tag in soup.find_all(True)]
from collections import Counter
print(Counter(tags).most_common(20))
```

```python
from bs4 import BeautifulSoup

with open("../data/raw/apple/10k_latest.html", "rb") as f:
    soup = BeautifulSoup(f, "html.parser")

# Look at span tags that might be section headers
# Short text, likely bold or styled
for span in soup.find_all("span"):
    text = span.get_text(strip=True)
    if 10 < len(text) < 100:
        style = span.get("style", "")
        if any(x in style.lower() for x in ["bold", "font-weight"]):
            print(repr(text[:80]), "|", style[:80])
```
